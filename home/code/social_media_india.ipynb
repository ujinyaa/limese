```python
import numpy as np  
import pandas as pd    
import matplotlib.pyplot as plt
import seaborn as sns

file_path = r"Social Media Usage India.csv"
data = pd.read_csv(file_path)
data.head()
```

```python

age_data_sorted = data.sort_values(by='Age', ascending=True).copy()

age_group_labels = ['age group: 10s', 'age group: 20s', 'age group: 30s', 'age group: 40s', 'age group over 50']
age_data_sorted['Age Group'] = pd.cut(age_data_sorted['Age'], bins=[0, 19, 29, 39, 49, 100], labels=age_group_labels, right=True)

age_group_data = {label: age_data_sorted[age_data_sorted['Age Group'] == label].head(5) for label in age_group_labels}


for label in age_group_labels:
    print(f"{label} Data:")
    display(age_group_data[label])


```

```python
data['Time Spent on WhatsApp in last week (in minutes)'] = pd.to_numeric(data['Time Spent on WhatsApp in last week (in minutes)'], errors='coerce')
data['Time Spent on WhatsApp in last weekend (in minutes)'] = pd.to_numeric(data['Time Spent on WhatsApp in last weekend (in minutes)'], errors='coerce')
data['Total Social Media Usage'] = pd.to_numeric(data['Total Social Media Usage'], errors='coerce')
data['Total Week Usage'] = pd.to_numeric(data['Total Week Usage'], errors='coerce')
data['Total Weekend Usage'] = pd.to_numeric(data['Total Weekend Usage'], errors='coerce')
data['Total WhatsApp Usage'] = pd.to_numeric(data['Total WhatsApp Usage'], errors='coerce')


analysis_results = {}

# 1. Age: Minimum, Maximum, Unique age count
min_age = data['Age'].min()
max_age = data['Age'].max()
unique_age_count = data['Age'].nunique()
analysis_results['Age'] = {
    'Min': min_age,
    'Max': max_age,
    'Unique Age Count': unique_age_count
}

# 2. City: Unique city count
unique_city_count = data['City'].nunique()
analysis_results['City'] = f"Total unique cities: {unique_city_count}"

# 3. Current Status: Count of unique types
unique_current_status_count = data['Current Status'].nunique()
analysis_results['Current Status'] = f"Total unique statuses: {unique_current_status_count}"

# 4. Gender: Unique values as list
unique_genders = data['Gender'].unique().tolist()
analysis_results['Gender'] = unique_genders

# 5. Highest Education: Count of unique types
unique_highest_education_count = data['Highest Education'].nunique()
analysis_results['Highest Education'] = f"Total unique education levels: {unique_highest_education_count}"

# 6. Time Spent on WhatsApp in last week (in minutes): Min, Max, Bins of 1000, 2000, ..., 6000+
min_whatsapp_week = data['Time Spent on WhatsApp in last week (in minutes)'].min()
max_whatsapp_week = data['Time Spent on WhatsApp in last week (in minutes)'].max()
bins_whatsapp_week = pd.cut(data['Time Spent on WhatsApp in last week (in minutes)'], bins=[0, 1000, 2000, 3000, 4000, 5000, 6000, float('inf')]).value_counts(sort=False)
analysis_results['Time Spent on WhatsApp in last week (in minutes)'] = {
    'Min': min_whatsapp_week,
    'Max': max_whatsapp_week,
    'Bins': bins_whatsapp_week
}

# 7. Time Spent on WhatsApp in last weekend (in minutes): Min, Max, Bins of 500, 1000, ..., 2000+
min_whatsapp_weekend = data['Time Spent on WhatsApp in last weekend (in minutes)'].min()
max_whatsapp_weekend = data['Time Spent on WhatsApp in last weekend (in minutes)'].max()
bins_whatsapp_weekend = pd.cut(data['Time Spent on WhatsApp in last weekend (in minutes)'], bins=[0, 500, 1000, 1500, 2000, float('inf')]).value_counts(sort=False)
analysis_results['Time Spent on WhatsApp in last weekend (in minutes)'] = {
    'Min': min_whatsapp_weekend,
    'Max': max_whatsapp_weekend,
    'Bins': bins_whatsapp_weekend
}

# 8. Total Social Media Usage: Min, Max, Bins of 1000, 5000, 7000, ..., 15000+
min_social_media_usage = data['Total Social Media Usage'].min()
max_social_media_usage = data['Total Social Media Usage'].max()
bins_social_media_usage = pd.cut(data['Total Social Media Usage'], bins=[0, 1000, 5000, 7000, 10000, 15000, float('inf')]).value_counts(sort=False)
analysis_results['Total Social Media Usage'] = {
    'Min': min_social_media_usage,
    'Max': max_social_media_usage,
    'Bins': bins_social_media_usage
}

# 9. Total Week Usage: Min, Max, Bins of 1000, 5000, 7000, ..., 15000+
min_week_usage = data['Total Week Usage'].min()
max_week_usage = data['Total Week Usage'].max()
bins_week_usage = pd.cut(data['Total Week Usage'], bins=[0, 1000, 5000, 7000, 10000, 15000, float('inf')]).value_counts(sort=False)
analysis_results['Total Week Usage'] = {
    'Min': min_week_usage,
    'Max': max_week_usage,
    'Bins': bins_week_usage
}

# 10. Total Weekend Usage: Min, Max, Bins of 500, 1000, 1500, 2000+
min_weekend_usage = data['Total Weekend Usage'].min()
max_weekend_usage = data['Total Weekend Usage'].max()
bins_weekend_usage = pd.cut(data['Total Weekend Usage'], bins=[0, 500, 1000, 1500, 2000, float('inf')]).value_counts(sort=False)
analysis_results['Total Weekend Usage'] = {
    'Min': min_weekend_usage,
    'Max': max_weekend_usage,
    'Bins': bins_weekend_usage
}

# 11. Total WhatsApp Usage: Min, Max, Bins of 500, 1000, 2000, ..., 7000+
min_whatsapp_usage = data['Total WhatsApp Usage'].min()
max_whatsapp_usage = data['Total WhatsApp Usage'].max()
bins_whatsapp_usage = pd.cut(data['Total WhatsApp Usage'], bins=[0, 500, 1000, 2000, 3000, 5000, 7000, float('inf')]).value_counts(sort=False)
analysis_results['Total WhatsApp Usage'] = {
    'Min': min_whatsapp_usage,
    'Max': max_whatsapp_usage,
    'Bins': bins_whatsapp_usage
}

analysis_results
```

```python
selected_columns = [
    'Age', 'Current Status', 'Gender', 'Highest Education', 
    'Time Spent on WhatsApp in last week (in minutes)', 
    'Time Spent on WhatsApp in last weekend (in minutes)', 
    'Total Social Media Usage', 'Total Week Usage', 
    'Total Weekend Usage', 'Total WhatsApp Usage'
]
labels = ['age group: 10s', 'age group: 20s', 'age group: 30s', 'age group: 40s', 'age group over 50']
data['Age Group'] = pd.qcut(data['Age'], q=5, labels=labels)

for column in selected_columns:
    plt.figure(figsize=(8, 6))

    if data[column].dtype in ['int64', 'float64']: 
        sns.boxplot(x='Age Group', y=column, data=data) 
        plt.title(f'Distribution of {column} by Age Group')
    else:  # For categorical columns
        sns.countplot(x='Age Group', hue=column, data=data)  
        plt.title(f'Distribution of {column} by Age Group')
    
    plt.xlabel("Age Group")
    plt.ylabel('Frequency / Value')
    plt.xticks(rotation=45) 
    plt.tight_layout()
    plt.show()
```

```python
top_20_city_list = data['City'].value_counts().nlargest(20).index
filtered_data = data[data['City'].isin(top_20_city_list)]

selected_columns = ['City']



for column in selected_columns:
    plt.figure(figsize=(10, 8))
    
    if filtered_data[column].dtype in ['int64', 'float64']:
        sns.boxplot(x='Age Group', y=column, data=filtered_data)
        plt.title(f'Distribution of {column} by Age Group')
        plt.ylabel('Value')
    else:
        sns.countplot(x='Age Group', hue=column, data=filtered_data)
        plt.title(f'Distribution of {column} by Age Group')
        plt.ylabel('Frequency')
    
    plt.xlabel("Age Group")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

palette = sns.color_palette("Set3")
```

```python
data['Time Spent on WhatsApp in last weekend (in minutes)'] = pd.to_numeric(
    data['Time Spent on WhatsApp in last weekend (in minutes)'].replace(',', '', regex=True), errors='coerce')
data['Time Spent on WhatsApp in last week (in minutes)'] = pd.to_numeric(
    data['Time Spent on WhatsApp in last week (in minutes)'].replace(',', '', regex=True), errors='coerce')
data['Total Social Media Usage'] = pd.to_numeric(
    data['Total Social Media Usage'].replace(',', '', regex=True), errors='coerce')
data['Total WhatsApp Usage'] = pd.to_numeric(
    data['Total WhatsApp Usage'].replace(',', '', regex=True), errors='coerce')
data['Total Facebook Usage'] = pd.to_numeric(
    data['Total Facebook Usage'].replace(',', '', regex=True), errors='coerce')
data['Total Instagram Usage'] = pd.to_numeric(
    data['Total Instagram Usage'].replace(',', '', regex=True), errors='coerce')
```

```python
weekend_week_correlation = data[['Time Spent on WhatsApp in last weekend (in minutes)', 
                                 'Time Spent on WhatsApp in last week (in minutes)']].corr().iloc[0, 1]

weekend_week_matrix_data = {
    'Metric': ['Correlation'],
    'Value': [weekend_week_correlation]
}
weekend_week_matrix = pd.DataFrame(weekend_week_matrix_data)

# Displaying the correlation matrix as a heatmap
plt.figure(figsize=(6, 4))
correlation_matrix = data[['Time Spent on WhatsApp in last weekend (in minutes)', 
                           'Time Spent on WhatsApp in last week (in minutes)']].corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', cbar=True, fmt=".2f",
            xticklabels=['Weekend WhatsApp Usage', 'Week WhatsApp Usage'],
            yticklabels=['Weekend WhatsApp Usage', 'Week WhatsApp Usage'])

plt.title("Correlation Matrix: WhatsApp Usage (Weekend vs. Week)")
plt.tight_layout()
plt.show()
```

```python
data['WhatsApp to Social Media Ratio (%)'] = (data['Total WhatsApp Usage'] / data['Total Social Media Usage']) * 100


whatsapp_social_media_summary = {
    'Metric': ['Mean (%)', 'Standard Deviation (%)', 'Minimum (%)', 'Maximum (%)'],
    'WhatsApp to Social Media Ratio': [
        data['WhatsApp to Social Media Ratio (%)'].mean(),
        data['WhatsApp to Social Media Ratio (%)'].std(),
        data['WhatsApp to Social Media Ratio (%)'].min(),
        data['WhatsApp to Social Media Ratio (%)'].max()
    ]
}
whatsapp_social_media_ratio_matrix = pd.DataFrame(whatsapp_social_media_summary)

plt.figure(figsize=(8, 2))
sns.heatmap(whatsapp_social_media_ratio_matrix.set_index('Metric'), annot=True, fmt=".2f", cmap="Blues", cbar=False)

plt.title("WhatsApp to Social Media Usage Ratio (%) Matrix")
plt.yticks(rotation=0)
plt.show()
```

```python
whatsapp_facebook_instagram_matrix_data = {
    'Metric': ['Mean (%)', 'Standard Deviation (%)', 'Minimum (%)', 'Maximum (%)'],
    'WhatsApp to Facebook Ratio': [
        data['Total WhatsApp Usage'].mean() / data['Total Facebook Usage'].mean() * 100,
        data['Total WhatsApp Usage'].std() / data['Total Facebook Usage'].std() * 100,
        data['Total WhatsApp Usage'].min() / data['Total Facebook Usage'].max() * 100,
        data['Total WhatsApp Usage'].max() / data['Total Facebook Usage'].min() * 100
    ],
    'WhatsApp to Instagram Ratio': [
        data['Total WhatsApp Usage'].mean() / data['Total Instagram Usage'].mean() * 100,
        data['Total WhatsApp Usage'].std() / data['Total Instagram Usage'].std() * 100,
        data['Total WhatsApp Usage'].min() / data['Total Instagram Usage'].max() * 100,
        data['Total WhatsApp Usage'].max() / data['Total Instagram Usage'].min() * 100
    ]
}
whatsapp_facebook_instagram_matrix = pd.DataFrame(whatsapp_facebook_instagram_matrix_data)

plt.figure(figsize=(10, 3))
sns.heatmap(whatsapp_facebook_instagram_matrix.set_index('Metric'), annot=True, fmt=".2f", cmap="Oranges", cbar=False)

plt.title("WhatsApp to Facebook and Instagram Usage Ratio (%) Matrix")
plt.yticks(rotation=0)
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import seaborn as sns
import pandas as pd

palette = sns.color_palette("Set3")

for age_group in age_group_labels:
    
    age_group_data = filtered_data[filtered_data['Age Group'] == age_group]
    
    plt.figure(figsize=(12, 10))
    map = Basemap(width=1200000, height=900000, projection='lcc', resolution='l',
                  llcrnrlon=67, llcrnrlat=5, urcrnrlon=99, urcrnrlat=37, lat_0=28, lon_0=77)
    
    map.drawmapboundary()
    map.drawcountries()
    map.drawcoastlines()
    
    x, y = map(age_group_data['Longitude'].values, age_group_data['Latitude'].values)
    
    plt.scatter(x, y, s=100, c=sns.color_palette('Set3', len(age_group_data)), marker="o", alpha=0.7)
    
    for city, xpt, ypt in zip(age_group_data['City'], x, y):
        plt.text(xpt + 5000, ypt + 5000, city, fontsize=10, fontweight='bold')
    
    plt.title(f'Major Cities in India by Latitude and Longitude (Age Group: {age_group})')
    plt.tight_layout()
    plt.show()
```
