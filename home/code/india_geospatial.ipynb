```python
import pandas as pd
import numpy as np
file_path = r"cities_r2.csv"
data = pd.read_csv(file_path)
data.head()
```

```python
data.describe()
```

Splitting location into lattitude and longitude for better and confortable visualization

```python
populated_cities = data.sort_values(by='population_total',ascending=False)
populated_cities.head(5)
```

```python
data['latitude'] = data['location'].apply(lambda x: x.split(',')[0])
data['longitude'] = data['location'].apply(lambda x: x.split(',')[1])
```

```python
populated_cities['latitude'] = data.loc[populated_cities.index, 'latitude']
populated_cities['longitude'] = data.loc[populated_cities.index, 'longitude']

populated_cities.head(5)
```

```python
from IPython.display import display
female_cities = data.sort_values(by='sex_ratio', ascending=False)[['name_of_city', 'population_total', 'population_female', 'sex_ratio']]
display(female_cities.head())  
```

```python
female_cities['latitude'] = data.loc[female_cities.index, 'latitude']
female_cities['longitude'] = data.loc[female_cities.index, 'longitude']

female_cities.head(5)
```

```python
data['city_literatability'] = data['literates_total'] / data['total_graduates'] 
city_literatability= data.sort_values(by='city_literatability', ascending=False)[
    ['name_of_city', 'city_literatability', 'literates_total', 'effective_literacy_rate_total', 'total_graduates']
]
city_literatability['latitude'] = data.loc[city_literatability.index, 'latitude']
city_literatability['longitude'] = data.loc[city_literatability.index, 'longitude']

city_literatability.head(5)
```

```python
data['female_literatability'] = data['literates_female'] / data['female_graduates']

female_literatability = data.sort_values(by='female_literatability', ascending=False)
[['name_of_city', 'female_literatability', 'literates_female', 'female_graduates', 'effective_literacy_rate_female', 'latitude', 'longitude']]

female_literatability['latitude'] = data.loc[female_literatability.index, 'latitude']
female_literatability['longitude'] = data.loc[female_literatability.index, 'longitude']

female_literatability.head(5)
```

```python
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.dates import date2num


from mpl_toolkits.basemap import Basemap
import matplotlib.cm as cm
```

```python
plt.figure(figsize=(20, 10)) 

plt.subplot(1, 2, 1) 
map = Basemap(width=1200000, height=900000, projection='lcc', resolution='l',
              llcrnrlon=67, llcrnrlat=5, urcrnrlon=99, urcrnrlat=37, lat_0=28, lon_0=77)

map.drawmapboundary()
map.drawcountries()
map.drawcoastlines()

longitude = np.array(populated_cities['longitude'])
latitude = np.array(populated_cities['latitude'])
total_population = np.array(populated_cities['population_total'])
city_name = np.array(populated_cities['name_of_city'])

x, y = map(longitude, latitude)
population_sizes = populated_cities["population_total"].head(50).apply(lambda x: int(x / 1000)) 
plt.scatter(x[:50], y[:50], s=population_sizes / 3, marker="o", c=population_sizes, cmap=cm.Dark2, alpha=0.8)  

for ncs, xpt, ypt in zip(city_name, x[:50], y[:50]):
    plt.text(xpt + 6000, ypt + 3000, ncs, fontsize=10, fontweight='bold')

plt.title('Top 50 Populated Cities in India', fontsize=20)


plt.subplot(1, 2, 2)

map = Basemap(width=1200000, height=900000, projection='lcc', resolution='l',
              llcrnrlon=67, llcrnrlat=5, urcrnrlon=99, urcrnrlat=37, lat_0=28, lon_0=77)

map.drawmapboundary()
map.drawcountries()
map.drawcoastlines()

longitude_female = np.array(female_cities['longitude'])
latitude_female = np.array(female_cities['latitude'])
female_population = np.array(female_cities['population_female'])
city_name_female = np.array(female_cities['name_of_city'])

x_female, y_female = map(longitude_female, latitude_female)
population_female_sizes = female_cities["population_female"].head(50).apply(lambda x: int(x / 1000)) 
plt.scatter(x_female[:50], y_female[:50], s=population_female_sizes * 2, marker="o", c=population_female_sizes, cmap=cm.Dark2, alpha=0.8)

for ncs, xpt, ypt in zip(city_name_female, x_female[:50], y_female[:50]):
    plt.text(xpt + 6000, ypt + 3000, ncs, fontsize=10, fontweight='bold')

plt.title('Top 50 Female Dominated Cities in India', fontsize=20)


plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20, 10))

plt.subplot(1, 2, 1)
map = Basemap(width=1200000, height=900000, projection='lcc', resolution='l',
              llcrnrlon=67, llcrnrlat=5, urcrnrlon=99, urcrnrlat=37, lat_0=28, lon_0=77)

map.drawmapboundary()
map.drawcountries()
map.drawcoastlines()

longitude = np.array(city_literatability['longitude'].values)  # 데이터프레임의 열을 numpy 배열로 변환
latitude = np.array(city_literatability['latitude'].values)
city_literacy_rate = np.array(city_literatability['city_literatability'].values)  # 도시 문해력 비율
city_name = np.array(city_literatability['name_of_city'].values)

x, y = map(longitude, latitude)
normalized_sizes = (city_literacy_rate[:50] - city_literacy_rate[:50].min()) / (city_literacy_rate[:50].max() - city_literacy_rate[:50].min())  # 정규화
min_marker_size = 80
max_marker_size = 1000
population_sizes = min_marker_size + (normalized_sizes * (max_marker_size - min_marker_size)).astype(int)

plt.scatter(x[:50], y[:50], s=population_sizes, marker="o", c=population_sizes, cmap=cm.Dark2, alpha=0.8)

for ncs, xpt, ypt in zip(city_name, x[:50], y[:50]):
    plt.text(xpt + 6000, ypt + 3000, ncs, fontsize=10, fontweight='bold')

plt.title('Top 50 Cities by Total Literability in India', fontsize=20)


plt.subplot(1, 2, 2)
map = Basemap(width=1200000, height=900000, projection='lcc', resolution='l',
              llcrnrlon=67, llcrnrlat=5, urcrnrlon=99, urcrnrlat=37, lat_0=28, lon_0=77)

map.drawmapboundary()
map.drawcountries()
map.drawcoastlines()

longitude_female = np.array(female_literatability['longitude'].values)
latitude_female = np.array(female_literatability['latitude'].values)
female_literacy_rate = np.array(female_literatability['female_literatability'].values)  # 여성 문해력 비율
city_name_female = np.array(female_literatability['name_of_city'].values)

x_female, y_female = map(longitude_female, latitude_female)
population_female_sizes = (female_literacy_rate[:50] / female_literacy_rate[:50].max() * 500).astype(int)
plt.scatter(x_female[:50], y_female[:50], s=population_female_sizes, marker="o", c=population_female_sizes, cmap=cm.Dark2, alpha=0.8)

for ncs, xpt, ypt in zip(city_name_female, x_female[:50], y_female[:50]):
    plt.text(xpt + 6000, ypt + 3000, ncs, fontsize=10, fontweight='bold')

plt.title('Top 50 Cities by Female Literability in India', fontsize=20)

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import display

Top250_city_sets = {
    'populated_cities': set(populated_cities['name_of_city'].head(250)),
    'female_cities': set(female_cities['name_of_city'].head(250)),
    'city_literatability': set(city_literatability['name_of_city'].head(250)),
    'female_literatability': set(female_literatability['name_of_city'].head(250))
}


Top150_city_sets = {
    'Top150_populated_cities': set(populated_cities['name_of_city'].head(150)),
    'Top150_female_cities': set(female_cities['name_of_city'].head(150)),
    'Top150_city_literatability': set(city_literatability['name_of_city'].head(150)),
    'Top150_female_literatability': set(female_literatability['name_of_city'].head(150))
}

Top250_city_sets_counts = pd.DataFrame(
    {
        key1: {
            key2: len(Top250_city_sets[key1].intersection(Top250_city_sets[key2]))
            for key2 in Top250_city_sets
        }
        for key1 in Top250_city_sets
    }
)

Top150_city_sets_counts = pd.DataFrame(
    {
        key1: {
            key2: len(Top150_city_sets[key1].intersection(Top150_city_sets[key2]))
            for key2 in Top150_city_sets
        }
        for key1 in Top150_city_sets
    }
)

display(Top250_city_sets_counts)
display(Top150_city_sets_counts)  
```

```python
corr_matrix_top250 = pd.DataFrame(
    {
        key1: {key2: len(Top250_city_sets[key1].intersection(Top250_city_sets[key2])) / ((len(Top250_city_sets[key1]) + len(Top250_city_sets[key2])) / 2)
               for key2 in Top250_city_sets}
        for key1 in Top250_city_sets
    }
)


corr_matrix_top150 = pd.DataFrame(
    {
        key1: {key2: len(Top150_city_sets[key1].intersection(Top150_city_sets[key2])) / ((len(Top150_city_sets[key1]) + len(Top150_city_sets[key2])) / 2)
               for key2 in Top150_city_sets}
        for key1 in Top150_city_sets
    }
)

fig, axes = plt.subplots(1, 2, figsize=(18, 8))


sns.heatmap(corr_matrix_top250, annot=True, cmap='coolwarm', linewidths=0.5, vmin=-1, vmax=1, ax=axes[0])
axes[0].set_title('Correlation Matrix of Top 250 Cities Sets')

sns.heatmap(corr_matrix_top150, annot=True, cmap='coolwarm', linewidths=0.5, vmin=-1, vmax=1, ax=axes[1])
axes[1].set_title('Correlation Matrix of Top 150 Cities Sets')

plt.tight_layout()
plt.show()
```
